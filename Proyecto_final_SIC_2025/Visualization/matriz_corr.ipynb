#### Bloque de codigo para generar la matriz de correlación de las variables de interés

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Datasets/final_data.csv")
```

```python
variables = {
    'surface_pressure': 'Presión',
    'surface_net_solar_radiation_sum': 'Radiación Solar',
    'total_precipitation_sum': 'Precipitación',
    'temperature_C': 'Temperatura',
    'aerosol': 'Aerosoles',
    'cloud_cover': 'Cobertura Nubes',
    'humidity': 'Humedad Relativa',
    'wind_speed': 'Vel. Viento',
    'wind_dir': 'Dir. Viento',
    'elevation': 'Elevación'
}

df_focus = df[variables.keys()]
```

```python
# Se calcula la matriz de correlación
corr = df_focus.corr()
corr = corr.rename(index=variables, columns=variables)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(
    corr,
    annot=True,
    cmap='Spectral',
    fmt='.2f',
    linewidths=0.5,
    cbar_kws={"shrink":0.8}
)
plt.title("Matriz de correlación", fontsize=16, pad=20)
plt.show()
```
